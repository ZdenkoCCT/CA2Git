## Data Cleaning

```python
#import  os libraries. Also importing warnings to disbale Python warning messages
import os
import warnings
warnings.filterwarnings('ignore')
files_list =os.listdir('.')
files_list
```

```python
#import pandas to load the data into the dataframe
import pandas as pd
```

```python
df_irl1= pd.read_csv(r'Average price 1970 -2001.csv')
#seting columns dispay option to 'max'- to see all columns in dataset
pd.set_option('display.max.columns', None)
df_irl1
```

```python
#index column and select only beef related products from the list
df_irl1 = df_irl1.set_index(['Consumer Item'])
df_irl1 =df_irl1.loc[df_irl1.index.isin(['Round steak per kg.','Sirloin steak per kg.','Rib steak per kg.','Leg beef (boneless) per kg.','Brisket pot roast per kg.'])]
df_irl1.shape
```

```python
df_irl2= pd.read_csv(r'Average price 2001-2011.csv')
#seting columns dispay option to 'max'- to see all columns in dataset
pd.set_option('display.max.columns', None)
df_irl2
```

```python
#index column and select only beef related products from the list
df_irl2 = df_irl2.set_index(['Consumer Item'])
df_irl2 =df_irl2.iloc[df_irl2.index.isin(['Round steak per kg.','Sirloin steak per kg.','Rib steak per kg.','Leg beef (boneless) per kg.','Brisket pot roast per kg.'])]
df_irl2.shape
```
